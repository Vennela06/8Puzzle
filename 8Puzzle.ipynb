```python
# Prompt for start and goal states
startState = []
# Creating output file
outFile = open("8PuzzleOutput.txt", "w")
print("Enter the start state (0-8, 0 represents the blank position) for 8-puzzle:")
for _ in range(3):
    row = list(map(int, input().split()))
    startState.append(row)

# Verify if the start state is valid
flatStartState = [num for sublist in startState for num in sublist]
validStartState = set(flatStartState) == set(range(9)) and len(flatStartState) == 9

while not validStartState:
    print("Invalid start state. Please enter a valid start state:")
    startState = []
    for _ in range(3):
        row = list(map(int, input().split()))
        startState.append(row)
    flatStartState = [num for sublist in startState for num in sublist]
    validStartState = set(flatStartState) == set(range(9)) and len(flatStartState) == 9
print(startState)
outFile.write("Start state: "+ str(startState))

goalState = []

print("Enter the goal state (0-8, 0 represents the blank position) for 8-puzzle:")
for _ in range(3):
    row = list(map(int, input().split()))
    goalState.append(row)

# Verify if the goal state is valid
flatGoalState = [num for sublist in goalState for num in sublist]
validGoalState = set(flatGoalState) == set(range(9)) and len(flatGoalState) == 9

while not validGoalState:
    print("Invalid goal state. Please enter a valid goal state:")
    goalState = []
    for _ in range(3):
        row = list(map(int, input().split()))
        goalState.append(row)
    flatGoalState = [num for sublist in goalState for num in sublist]
    validGoalState = set(flatGoalState) == set(range(9)) and len(flatGoalState) == 9
print(goalState)
outFile.write("\nGoal state: "+ str(goalState))
outFile.close()
```

```python
# function to get possible moves from the current state
def getPossibleMoves(state):
    moves = []
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                if i > 0:
                    newState = [row[:] for row in state]
                    newState[i][j] = newState[i - 1][j]
                    newState[i - 1][j] = 0
                    moves.append(newState)
                if j > 0:
                    newState = [row[:] for row in state]
                    newState[i][j] = newState[i][j - 1]
                    newState[i][j - 1] = 0
                    moves.append(newState)
                if i < 2:
                    newState = [row[:] for row in state]
                    newState[i][j] = newState[i + 1][j]
                    newState[i + 1][j] = 0
                    moves.append(newState)    
                if j < 2:
                    newState = [row[:] for row in state]
                    newState[i][j] = newState[i][j + 1]
                    newState[i][j + 1] = 0
                    moves.append(newState)
    return moves
```

```python
#function to find index in goal state
def findIndex(goal, tile):
    row = col = None
    for i in range(3):
        for j in range(3):
            if goal[i][j] == tile:
                row = i
                col = j
                break
        if row is not None and col is not None:
            break
    return row, col 
```

```python
# Manhattan distance heuristic function
def findManhattanDistance(state, goal):
    manhattanDistance = 0
    for i in range(3):
        for j in range(3):
            tile = state[i][j]
            if tile != 0:
                # Calling findIndex function to get row and column index of current tile in goal state
                row, col = findIndex(goal,tile)
                manhattanDistance += abs(i - row) + abs(j - col)
    return manhattanDistance
```

```python
# function to implement A* search algorithm
import heapq
def aStarSearch(start, goal):
    # Priority queue for the open states
    openStates = [(findManhattanDistance(start, goal), start)]
    heapq.heapify(openStates)

    # Dictionary to store the parent state for each state
    parents = {}
    
    # Dictionary to keep track of the g-scores for each state
    gSocres = {tuple(map(tuple, start)): 0}

    while openStates:
        _, currentState = heapq.heappop(openStates)

        if currentState == goal:
            # Reconstruct the path from goal to start
            path = []
            while currentState != start:
                path.append(currentState)
                currentState = parents[tuple(map(tuple, currentState))]
            path.append(start)
            path.reverse()
            return path

        for nextState in getPossibleMoves(currentState):
            newScore = gSocres[tuple(map(tuple, currentState))] + 1
            if tuple(map(tuple, nextState)) not in gSocres or newScore < gSocres[tuple(map(tuple, nextState))]:
                gSocres[tuple(map(tuple, nextState))] = newScore
                score = newScore + findManhattanDistance(nextState, goal)
                heapq.heappush(openStates, (score, nextState))
                parents[tuple(map(tuple, nextState))] = currentState

    return None
```

```python
# Calling aStarSearch function to solve the puzzle using A* Search algorithm
path = aStarSearch(startState, goalState)
# Writing to output file
outFile = open("8PuzzleOutput.txt","a")
if path is None:
    print("\nNo solution found for the given puzzle")
    outFile.write("\nNo solution found for the given puzzle")
else:
    print("\nSequence of moves to solve the 8-puzzle using A* search:")
    outFile.write("\nSequence of moves to solve the 8-puzzle using A* search:")
    
    for move in path:
        for row in move:
            print(row)
            outFile.write("\n"+str(row))
        print()
        outFile.write("\n")
outFile.close()
```
